## Parameter selection 

```python
# Read in ichaway files 

# read in SA files
sa_pulse_df = pd.read_csv("./ich_SA_runs/ribbit_scores_pulse_rate_range/ribbit_scores2.csv")
sa_pulse_df['date']=pd.to_datetime(sa_pulse_df['date'])


# read in SA files
sa_noise_df = pd.read_csv("./ich_SA_runs/ribbit_scores_noise_bands/ribbit_scores1.csv")
sa_noise_df['date']=pd.to_datetime(sa_noise_df['date'])

# read in SA files
sa_duration_df = pd.read_csv("./ich_SA_runs/ribbit_scores_clip_duration/ribbit_scores1.csv")
sa_duration_df['date']=pd.to_datetime(sa_duration_df['date'])

# read in SA files
sa_signal_df = pd.read_csv("./ich_SA_runs/ribbit_scores_signal_band/ribbit_scores2.csv")
sa_signal_df['date']=pd.to_datetime(sa_signal_df['date'])

# read in SA files
sa_window_df = pd.read_csv("./ich_SA_runs/ribbit_scores_window_samples_base/ribbit_scores1.csv")
sa_window_df['date']=pd.to_datetime(sa_window_df['date'])

# read in SA files - this has noise band [1100,1200] instead of base noise band
sa_window_df2 = pd.read_csv("./ich_SA_runs/ribbit_scores_window_samples/ribbit_scores1.csv")
sa_window_df2['date']=pd.to_datetime(sa_window_df['date'])

# read in SA files
sa_optimal_guess_df = pd.read_csv("./ich_SA_runs/ribbit_scores_optimal_guess/ribbit_scores0.csv")
sa_optimal_guess_df['date']=pd.to_datetime(sa_optimal_guess_df['date'])


# read in SA files
sa_optimal_guess_df2 = pd.read_csv("./ich_SA_runs/ribbit_scores_optimal_guess_ws400/ribbit_scores0.csv")
sa_optimal_guess_df2['date']=pd.to_datetime(sa_optimal_guess_df2['date'])



# these files were labeled incorrectly in the ichaway data - there are gopher frogs in them 
# logger 5a: 
#20150205_194700
#20150205_204700
#20150205_214700

# fix these mistakes
temp = '/Volumes/Expansion/Frog Call Project/Calling Data/Frog Calls/ichaway_2015/5a/20150205_'
incorrect_files =  [temp + '194700.wav', temp + '204700.wav', temp + '214700.wav']
sa_pulse_df.loc[sa_pulse_df['file_path'].isin(incorrect_files),'Lcapito'] = 1
sa_noise_df.loc[sa_noise_df['file_path'].isin(incorrect_files),'Lcapito'] = 1
sa_duration_df.loc[sa_duration_df['file_path'].isin(incorrect_files),'Lcapito'] = 1
sa_signal_df.loc[sa_signal_df['file_path'].isin(incorrect_files),'Lcapito'] = 1
sa_window_df.loc[sa_window_df['file_path'].isin(incorrect_files),'Lcapito'] = 1
sa_window_df2.loc[sa_window_df2['file_path'].isin(incorrect_files),'Lcapito'] = 1
sa_optimal_guess_df.loc[sa_optimal_guess_df['file_path'].isin(incorrect_files),'Lcapito'] = 1
sa_optimal_guess_df2.loc[sa_optimal_guess_df2['file_path'].isin(incorrect_files),'Lcapito'] = 1

# make 'Lcapito' columns categories not numerics
sa_pulse_df['Lcapito'] = sa_pulse_df['Lcapito'].astype('category')
sa_noise_df['Lcapito'] = sa_noise_df['Lcapito'].astype('category')
sa_duration_df['Lcapito'] = sa_duration_df['Lcapito'].astype('category')
sa_signal_df['Lcapito'] = sa_signal_df['Lcapito'].astype('category')
sa_window_df['Lcapito'] = sa_window_df['Lcapito'].astype('category')
sa_window_df2['Lcapito'] = sa_window_df2['Lcapito'].astype('category')
sa_optimal_guess_df['Lcapito'] = sa_optimal_guess_df['Lcapito'].astype('category')
sa_optimal_guess_df2['Lcapito'] = sa_optimal_guess_df2['Lcapito'].astype('category')

# create dataframe with base data from SA 
sa_base_df = sa_pulse_df.drop(columns = ['time_stamp0', 'score0', 'time_stamp1','score1']).rename(columns={'score2':'score', 'time_stamp2':'time_stamp'})
```

```python
# calculate precision recall: compare best parameter values for each parameter from the SAs
cutoffs = range(0,200)

ich_accuracy0 = calc_accuracy(sa_optimal_guess_df, cutoffs, score_col = 'score0')
ich_accuracy1 = calc_accuracy(sa_noise_df, cutoffs, score_col = "score0")
ich_accuracy2 = calc_accuracy(sa_signal_df, cutoffs, score_col = "score2")
ich_accuracy3 = calc_accuracy(sa_duration_df, cutoffs, score_col = "score0")
ich_accuracy4 = calc_accuracy(sa_pulse_df, cutoffs, score_col = "score1")
ich_accuracy5 = calc_accuracy(sa_optimal_guess_df2, cutoffs, score_col = "score0") 
ich_accuracy6 = calc_accuracy(sa_window_df, cutoffs, score_col = "score1") 
ich_accuracy_base = calc_accuracy(sa_pulse_df, cutoffs, score_col = "score2") # base vals
#ich_accuracy0 = calc_accuracy(sa_pulse_df[sa_pulse_df["pond"]!="7a"], cutoffs)

#pulse, noise, signal, duration, optimal guess


# plot precision and recall for pulse rate sa - looks like [12,28] is best of the options 

# best: 
# pulse: param1 (by a tad)
# noise: both better, param0 slightly better at first but param1 overtakes  - try 1100-1300 band? 
# signal: both better, param0/param2  better at first but param1 better later - probably go with param0/2 
# duration: pretty similar
# optimal guess: 
```
